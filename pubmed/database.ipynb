# Database

Import CSV file and build a SQLite database.  
Query database based on author's name.

```python
import sqlite3
import os
import pandas as pd

csv_fname = 'db/HIV.csv'
table_name = os.path.basename(csv_fname).split('.')[0].lower()
db_fname = csv_fname.replace('.csv', '.db')
```

Build database

```python
conn = sqlite3.connect(db_fname)
cursor = conn.cursor()
```

```python
# drop any existing table
cursor.execute('DROP TABLE IF EXISTS ' + table_name)

# ensure that first column is used as index;
# otherwise, another index column will be created
df = pd.read_csv(csv_fname, index_col=0)

# create sql table
df.to_sql(table_name, conn)
```

Query database

```python
author_name = input('Author name: ')
query = 'SELECT pmid, ti from {0} where au like \'%{1}%\';'.format(table_name, author_name)

for row in cursor.execute(query):
    print(row)
```
